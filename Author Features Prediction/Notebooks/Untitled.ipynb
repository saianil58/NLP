2. Preprocess rows of the “text” column (7.5 points)
a. Remove unwanted characters


```python
import re
def remove_special_characters(text, remove_digits=False):
    #Using regex
    pattern = r'[^a-zA-z0-9\s]' if not remove_digits else r'[^a-zA-z\s]'
    text = re.sub(pattern, '', text)
    return text
```

```python
df.text[0]
```

```python
remove_special_characters(df.text[0])
```

```python
df['clean_text']=df.apply(lambda row : remove_special_characters(row['text'], remove_digits=False), axis = 1)
```

```python
df.sample(5)
```

```python
df['clean_text'] = df['clean_text'].str.lower()
```

```python
df.sample(5)
```

```python
df['clean_text'] = df['clean_text'].str.strip()
```

```python
df.sample(5)
```

```python
from nltk.corpus import stopwords

cachedStopWords = stopwords.words("english")

def testFuncNew(text_to_clean):
    text = text_to_clean
    text = ' '.join([word for word in text.split() if word not in cachedStopWords])
    return text
```

```python
df['clean_text'][0]
```

```python
testFuncNew(df['clean_text'][0])
```

```python
df['clean_text']=df.apply(lambda row : testFuncNew(row['clean_text']), axis = 1)
```

```python
df.sample(2)
```

```python
cols = [ 'gender', 'age', 'topic', 'sign']
df['age'] = df['age'].astype('str')
df['labels'] = df[cols].values.tolist()
```

```python
df.sample(2)
```

Separate features and labels, and split the data into training and testing (5 points)

```python
from sklearn.model_selection import train_test_split
```

```python
# split X and y into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.clean_text, df.labels, random_state=2020)
```

Vectorize the features (5 points)


```python
# import and instantiate CountVectorizer (with the default parameters)
from sklearn.feature_extraction.text import CountVectorizer
cvect = CountVectorizer(ngram_range=(1,1))
```

```python
#Feed SMS data to CountVectorizer
cvect.fit(X_train)
```

```python
#Check the vocablury size
len(cvect.vocabulary_)
```

```python
#What is there in the vocabulary
cvect.vocabulary_
```

```python
cvect.get_feature_names()
```

```python
#Convert Training data into Count Vectors
X_train_ct = cvect.transform(X_train)
```

```python
X_test_ct = cvect.transform(X_test)
```

As we have noticed before, in this task each example can have multiple tags. To deal with
such kind of prediction, we need to transform labels in a binary form and the prediction will be
a mask of 0s and 1s. For this purpose, it is convenient to use MultiLabelBinarizer from sklearn
a. Convert your train and test labels using MultiLabelBinarizer

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
```

```python
df.labels
```

```python
mlb.fit(y_train)
```

```python
mlb.classes_
```

```python
y_train = mlb.transform(y_train)
```

```python
y_test = mlb.transform(y_test)
```

```python
y_train[0]
```

```python
import numpy as np
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf = OneVsRestClassifier(clf)
```

 Fit the classifier, make predictions and get the accuracy (5 points)

```python
clf.fit(X_train_ct, y_train)
```

```python
y_preds = clf.predict(X_test_ct)
```

```python
y_preds[0]
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_preds))
```

```python
from sklearn.metrics import f1_score
print(f1_score(y_test,y_preds,average='micro'))
```

```python
from sklearn.metrics import average_precision_score
print(average_precision_score(y_test,y_preds,average='micro'))
```

```python
from sklearn.metrics import recall_score
print(recall_score(y_test,y_preds,average='micro'))
```
